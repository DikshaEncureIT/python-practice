```python
from numpy import random

x = random.randint(100)
print(x)

x = random.rand(5)
print(x)
```

```python
# Generate Random Array
x=random.randint(100, size=(5))
print(x)

x = random.randint(100, size=(5, 5))
print(x)

x = random.rand(5)

print(x)

x = random.choice([3, 5, 7, 9])

print(x)
```

```python
x = random.choice([3, 5, 7, 9], size=(3, 5))

print(x)
```

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x)
```

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.4, 0.2], size=(10))

print(x)
```

```python
# Random Distribution
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.5, 0.1], size=(50))
print(x)
```

```python
# return a 2-D array with 3 rows, each containing 5 values.
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))
print(x)
```

```python
# shuffle example
from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr)
print(arr)
```

```python
# permutation example
arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr))
```
